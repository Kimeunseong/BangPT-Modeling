```python
import os
import numpy as np
import cv2
import matplotlib.pyplot as plt

path = './data/total_0511/2_images/images_dataset/'
img_list = []

for i in sorted(os.listdir(path)):
    if i != '.DS_Store':
        img_list.append(i)

# img_list
wrong_shape_list = []

for i in img_list:
    img = cv2.imread(path + i + '/exercise_00.jpg')
    if img.shape != (1280, 720, 3):
        print(f'폴더 {i} 이미지 shape이 다름!, shape = {img.shape}')
        wrong_shape_list.append(i)

# wrong_shape_list
print(len(wrong_shape_list))
```
